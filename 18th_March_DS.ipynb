1)Give numpy input for bar chart and line chart
   Shld include title, x and y label, legend, grid and marker


```python
import numpy as np
import matplotlib.pyplot as plt

# Create numpy data
x = np.array(['A', 'B', 'C', 'D', 'E'])
y = np.array([10, 25, 15, 30, 20])

# Create the plot
plt.figure(figsize=(10, 5))

# 1. Bar Chart
plt.bar(x, y, color='skyblue', alpha=0.6, label='Bar Data')

# 2. Line Chart (with markers)
plt.plot(x, y, color='red', marker='o', linestyle='-', linewidth=2, label='Line Data')

# Customization
plt.title('Comparison of Bar and Line Charts')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()  # Legend
plt.grid(True, linestyle='--', alpha=0.7)  # Grid

# Display
plt.show()
```

2)2-D array data with SubPlots and make Bar chat and Histrogram



```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Create varied 2-D data for a more dynamic look
np.random.seed(42)
data_2d = np.array([
    np.random.normal(50, 5, 200),
    np.random.normal(70, 15, 200),
    np.random.gamma(2, 10, 200) + 20,
    np.random.normal(40, 8, 200)
]).T

col_means = np.mean(data_2d, axis=0)
col_labels = ['Growth', 'Volatility', 'Output', 'Efficiency']
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']

# 2. Create Subplots with improved styling
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
plt.style.use('seaborn-v0_8-muted')

# --- Subplot 1: Stylized Bar Chart ---
ax1.bar(col_labels, col_means, color=colors, edgecolor='white', linewidth=1.5)
ax1.set_title('Performance Metrics (Column Means)', fontsize=14, fontweight='bold', pad=15)
ax1.set_ylabel('Average Value', fontsize=12)
ax1.set_ylim(0, 100)
ax1.grid(axis='y', linestyle=':', alpha=0.5)

# --- Subplot 2: Stacked Histogram ---
# Using a list of arrays for a stacked visualization
ax2.hist([data_2d[:, i] for i in range(4)], bins=25, stacked=True,
         label=col_labels, color=colors, alpha=0.8, edgecolor='grey')

ax2.set_title('Stacked Distribution Composition', fontsize=14, fontweight='bold', pad=15)
ax2.set_xlabel('Measured Value', fontsize=12)
ax2.set_ylabel('Cumulative Frequency', fontsize=12)
ax2.legend(frameon=True, facecolor='white')
ax2.grid(True, linestyle=':', alpha=0.5)

plt.tight_layout()
plt.show()
```
