### What is a DataFrame?

A `DataFrame` in pandas is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns). Think of it like a spreadsheet or a SQL table. It's the most commonly used pandas object and is ideal for storing and manipulating data.

### Accessing Data: `loc` and `iloc`

Pandas provides two primary methods to access data by label (`loc`) or by integer position (`iloc`).

*   **`loc` (label-based indexing):** This method is used for selecting data based on the *labels* of rows and columns. You can use it with a single label, a list of labels, a slice of labels, or a boolean array. It's inclusive of the end label when slicing.

*   **`iloc` (integer-location based indexing):** This method is used for selecting data based on the *integer positions* of rows and columns. It's similar to standard Python list slicing, where the start index is inclusive and the stop index is exclusive.

### Example Scenario: Employee Records DataFrame

Let's create a 7x5 DataFrame representing employee records and then demonstrate `loc`, `iloc`, and other operations.

```python
import pandas as pd

# Create a dictionary to hold our employee data
data = {
    'Employee ID': ['E001', 'E002', 'E003', 'E004', 'E005', 'E006', 'E007'],
    'Name': ['Alice Smith', 'Bob Johnson', 'Charlie Brown', 'Diana Prince', 'Eve Adams', 'Frank White', 'Grace Lee'],
    'Department': ['Sales', 'Marketing', 'Engineering', 'HR', 'Sales', 'Engineering', 'Marketing'],
    'Salary': [60000, 75000, 90000, 65000, 62000, 95000, 78000],
    'Hire Date': ['2020-01-15', '2019-03-20', '2021-06-01', '2018-11-01', '2022-02-10', '2020-07-01', '2021-09-01']
}

# Create the DataFrame
employees_df = pd.DataFrame(data)

# Convert 'Hire Date' to datetime objects for better handling
employees_df['Hire Date'] = pd.to_datetime(employees_df['Hire Date'])

print("Employee Records DataFrame (7 rows x 5 columns):")
display(employees_df)
```

### Using `loc` (Label-based Indexing)

`loc` allows you to select rows and columns by their labels.

```python
# Select a single row by its index label (e.g., the row with index 0)
print("\nSingle row with index label 0:")
display(employees_df.loc[0])

# Select multiple rows by their index labels (e.g., rows with index 1, 3, 5)
print("\nRows with index labels 1, 3, 5:")
display(employees_df.loc[[1, 3, 5]])

# Select rows using a slice of labels (inclusive of both start and end labels)
print("\nRows from index 2 to 4 (inclusive):")
display(employees_df.loc[2:4])

# Select specific columns for all rows
print("\n'Name' and 'Department' columns for all employees:")
display(employees_df.loc[:, ['Name', 'Department']])

# Select specific rows and specific columns
print("\n'Name' and 'Salary' for rows with index labels 0, 2, 6:")
display(employees_df.loc[[0, 2, 6], ['Name', 'Salary']])

# Select rows based on a condition
print("\nEmployees in the 'Engineering' department:")
display(employees_df.loc[employees_df['Department'] == 'Engineering'])

# Select 'Name' and 'Salary' for employees with Salary > 70000
print("\nNames and Salaries for employees with Salary > 70000:")
display(employees_df.loc[employees_df['Salary'] > 70000, ['Name', 'Salary']])
```

### Using `iloc` (Integer-location based Indexing)

`iloc` allows you to select rows and columns by their integer positions, starting from 0.

```python
# Select a single row by its integer position (e.g., the first row)
print("\nSingle row at integer position 0:")
display(employees_df.iloc[0])

# Select multiple rows by their integer positions (e.g., rows at position 1, 3, 5)
print("\nRows at integer positions 1, 3, 5:")
display(employees_df.iloc[[1, 3, 5]])

# Select rows using an integer slice (exclusive of the end index)
print("\nRows from position 2 up to (but not including) 5:")
display(employees_df.iloc[2:5])

# Select specific columns by their integer positions for all rows
print("\nColumns at integer positions 1 and 2 ('Name' and 'Department') for all employees:")
display(employees_df.iloc[:, [1, 2]])

# Select specific rows and specific columns by integer positions
print("\nRows at positions 0, 2, 6 and columns at positions 1, 3 ('Name', 'Salary'):")
display(employees_df.iloc[[0, 2, 6], [1, 3]])

# Select the last row
print("\nLast row using negative indexing:")
display(employees_df.iloc[-1])
```

### Common Operations with DataFrames

DataFrames are powerful and offer many operations for data manipulation and analysis.

```python
# 1. Getting information about the DataFrame
print("\nDataFrame Info:")
employees_df.info()

# 2. Descriptive statistics for numerical columns
print("\nDescriptive Statistics for Salary:")
display(employees_df['Salary'].describe())

# 3. Checking for null values
print("\nNull values per column:")
display(employees_df.isnull().sum())

# 4. Sorting data
print("\nDataFrame sorted by Salary (descending):")
display(employees_df.sort_values(by='Salary', ascending=False).head())

# 5. Grouping data
print("\nAverage Salary by Department:")
display(employees_df.groupby('Department')['Salary'].mean())

# 6. Adding a new column (e.g., 'Years at Company')
# Assuming today's date for calculation
from datetime import datetime
today = datetime.now()
employees_df['Years at Company'] = (today - employees_df['Hire Date']).dt.days / 365.25
print("\nDataFrame with 'Years at Company' column:")
display(employees_df.head())

# 7. Filtering data based on multiple conditions
print("\nEmployees in Sales with Salary > 60000:")
display(employees_df[(employees_df['Department'] == 'Sales') & (employees_df['Salary'] > 60000)])

# 8. Updating values (e.g., giving a salary raise to Marketing department)
employees_df.loc[employees_df['Department'] == 'Marketing', 'Salary'] *= 1.05 # 5% raise
print("\nSalaries updated for Marketing department (5% raise):")
display(employees_df[employees_df['Department'] == 'Marketing'])

# 9. Dropping a column
# Let's drop 'Years at Company' for demonstration, then add it back if needed
employees_df_no_years = employees_df.drop(columns=['Years at Company'])
print("\nDataFrame after dropping 'Years at Company' column:")
display(employees_df_no_years.head())
```
